```python
#entrando e logando
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

navegador = webdriver.Chrome()

navegador.get('https://blaze.com')

clicando_no_login = navegador.find_element(By.XPATH, '//*[@id="header"]/div/div[2]/div/div/div[1]/a')

clicando_no_login.click()
esperando_login = time.sleep(5)

email = navegador.find_element(By.XPATH, '//*[@id="auth-modal"]/div[2]/form/div[1]/div/input')

email.click()
email.send_keys('SEU EMAIL AQUI')

senha = navegador.find_element(By.XPATH, '//*[@id="auth-modal"]/div[2]/form/div[2]/div/input')

senha.click()
senha.send_keys('SUA SENHA AQUI')

botao_login = navegador.find_element(By.XPATH, '//*[@id="auth-modal"]/div[2]/form/div[4]/button')

botao_login.click()

esperando_carregar = time.sleep(5)
double = navegador.find_element(By.XPATH, '//*[@id="original-games"]/div/div[2]/div/a')

double.click()
contador = 0
contador_antes = 0
qtd_entradas = 0
tempo_execucao = 0
tempo_inicial = time.time()

#contando e verificando os numeros
while tempo_execucao < 18000:
 
    try:
        verificacao_time = navegador.find_element(By.XPATH, '//*[@id="roulette-timer"]/div/div[2]').text
        
        plv_verifica = verificacao_time.split()
        
        
        if (len(plv_verifica) > 0):

            pegar_dados = navegador.find_element(By.XPATH, '//*[@id="roulette-recent"]/div/div[1]/div[1]').text

            separando = pegar_dados.split()

            if (len(separando) >= 1):
                contador += 1
                print("Rodadas sem branco: {}".format(contador))
                print(separando)
                
                #padrao para entrada
                if (contador >= 5 and contador < 10):
                    
                    branco = navegador.find_element(By.XPATH, '//*[@id="roulette-controller"]/div[1]/div[2]/div[2]/div/div[2]')
                    branco.click()

                    colocando_valor = navegador.find_element(By.XPATH, '//*[@id="roulette-controller"]/div[1]/div[2]/div[1]/div/div[1]/input')
                    colocando_valor.send_keys('7') #valor da aposta
                    qtd_entradas += 1

                    time.sleep(2)
                    botao_aposta = navegador.find_element(By.XPATH, '//*[@id="roulette-controller"]/div[1]/div[3]/button')
                    botao_aposta.click()
                    print("entrada")
                
                                
            else:
                contador = 0
                print("DEU Branco")
                hora = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime())
                print(hora)

        #predendo para esperar o momento de fazer a aposta
        
        while(len(plv_verifica) > 0):

            verificacao_time = navegador.find_element(By.XPATH, '//*[@id="roulette-timer"]/div/div[2]').text

            plv_verifica = verificacao_time.split()
            
            #tempo_execucao = time.time() - tempo_inicial(deixe assim se quiser que rode durante um periodo nao determinado)
    except:
        time.sleep(1)

```
